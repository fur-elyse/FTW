```python
print('Path to Driver is '+filepath)
browser = webdriver.Chrome(executable_path=filepath)

list_philkotse = []
with open('C:\\Users\\Cobena\\Desktop/list_wpg.txt') as a:
    list_philkotse = a.readlines()
list_philkotse = [x.rstrip('\n') for x in list_philkotse] 
# print (list_philkotse)

df=pd.DataFrame()

for urls in list_philkotse:
#     df=pd.DataFrame()
    browser.get(urls)
    try:
        #reviews element
        title  = WebDriverWait(browser,5).until(
            EC.presence_of_element_located((By.XPATH,'//*[@id="title-detail"]'))).get_attribute('innerHTML')
        title=title.rstrip()
        big_div = WebDriverWait(browser,5).until(
            EC.presence_of_element_located((By.XPATH,'//*[@id="wrapper"]')))
        specs = big_div.find_elements_by_class_name('note-report')
        for spec in specs:
            try:
                price = spec.find_element_by_xpath('//*[@id="price_auto"]').get_attribute('innerHTML')
                brand1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[1]/span[1]').get_attribute('innerHTML')
                model1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[2]/span[1]').get_attribute('innerHTML')
                version1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[3]/span[1]').get_attribute('innerHTML')
                yr_of_man1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[4]/span[1]').get_attribute('innerHTML')
                condition1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[5]/span[1]').get_attribute('innerHTML')
                transmission1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[6]/span[1]').get_attribute('innerHTML')
                mileage1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[7]/span[1]').get_attribute('innerHTML')
                color1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[6]/span[1]').get_attribute('innerHTML')
                body_type1 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[7]/span[1]').get_attribute('innerHTML')
                brand2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[1]/span[2]').get_attribute('innerHTML')
                model2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[2]/span[2]').get_attribute('innerHTML')        
                version2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[3]/span[2]').get_attribute('innerHTML')        
                yr_of_man2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[4]/span[2]').get_attribute('innerHTML')
                condition2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[5]/span[2]').get_attribute('innerHTML')
                transmission2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[6]/span[2]').get_attribute('innerHTML')
                mileage2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[7]/span[2]').get_attribute('innerHTML')
                color2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[6]/span[2]').get_attribute('innerHTML')
                body_type2 = spec.find_element_by_xpath('//*[@id="wrapper"]/div/div/div[2]/div[2]/div[2]/div/ul/li[7]/span[2]').get_attribute('innerHTML')
                desc = WebDriverWait(browser,5).until(
                    EC.presence_of_element_located((By.XPATH,'//*[@id="wrapper"]/div/div/div[2]/div[4]/div[2]'))).get_attribute('innerHTML')
            except:
                print("Skipped",urls)
        new_entry = {"Title":title,"Price":price,"body_type1":body_type1,"body_type2":body_type2,"brand1":brand1,"brand2":brand2,"color1":color1,"color2":color2,"condition1":condition1,"condition2":condition2,"mileage1":mileage1,"mileage2":mileage2,"model1":model1,"model2":model2,"transmission1":transmission1,"transmission2":transmission2,"version1":version1,"version2":version2,"yr_of_man1":yr_of_man1,"yr_of_man2":yr_of_man2,"Description":desc}
        order_list = ["Title", "Price", "brand1","brand2","model1","model2","yr_of_man1","yr_of_man2","condition1","condition2","transmission1","transmission2","mileage1","mileage2","color1","color2","body_type1","body_type2","version1","version2","Description"]
        df = df.append(new_entry, ignore_index=True)
        df = df[order_list]

    except TimeoutException:
        print('failed to load element')

    try:
        savepath = "C:\\Users\\COBENA\\Desktop\\Learn\\Code\\FTW\\Capstone Idea\\Automotive\\Data Scraped\\philkotse_car_brand-specs.csv"
        count=1
        print(count)
        philkotse = df.to_csv(savepath, index=None, header=False)

    except Exception as e:
        print ("Export exception:", e)    

browser.quit()
```
