```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.common.keys import Keys
from time import sleep
from selenium.webdriver import ActionChains

import os
import pandas as pd
```

```python
##### Chromedriver #####
dirpath = os.getcwd()
filepath = dirpath + '/chromedriver'
print ('Path to Driver: ', filepath)
```

```python
browser = webdriver.Chrome(executable_path = filepath)
price_list=[1,2,3,4,5,6,8,9,10,11,12,13,15,19,20,21,22]
count=0

df=pd.DataFrame()
for num in price_list:
    browser.get('http://autosearchmanila.com/carpricelist.aspx?manID='+str(num))
    for x in range (1,100):
        try:
            title=browser.find_element_by_xpath('//*[@id="ctl00_ContentPlaceHolder1_ASPxRoundPanel_PriceListMan_ASPxCallbackPanel1_divPriceList"]/table/tbody/tr['+str(x)+']').text
            df=df.append({'title':title,'num':num},ignore_index=True)
            count+=1
            print(count)
        except:
            print('Skip')
    print('##################################')
try:
    savepath = "C:\\Users\\COBENA\\Desktop\\Learn\\Code\\FTW\\Capstone Idea\\Automotive\\Web Scraping\\retail_price.csv"
    philkotse = df.to_csv(savepath, index=None, header=False)

except Exception as e:
    print ("Export exception:", e)

print('done')
```
