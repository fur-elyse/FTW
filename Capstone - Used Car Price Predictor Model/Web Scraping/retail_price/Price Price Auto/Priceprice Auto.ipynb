## Scraping URLs from Price Price Auto

```python
browser = webdriver.Chrome(executable_path = filepath)
url = "https://ph.priceprice.com/cars/"
print ("##########################################")
print (url)

df=pd.DataFrame()
FinalPage=13

for x in range (1,FinalPage):
    if x == 1:
        url="https://ph.priceprice.com/cars/"
    else:
        url="https://ph.priceprice.com/cars/?page="+str(x)
    try:
        browser.get(url)
        car_list=browser.find_elements_by_class_name('pItem')
        for li in car_list:
            model=li.find_element_by_class_name('s-modelSum_link').get_attribute('href')
            print(model)
            df=df.append({'model':model},ignore_index=True)
    except Exception:
        print("End")
    print(x)
    
browser.quit()
print("Finished scraping. Browser closed")

######################################
#### Export the dataframe to CSV #####

try:
    savepath = "C:\\Users\COBENA\Desktop\Learn\Code\FTW\Capstone Idea\Automotive\Price Price Auto\ppa_urls.csv"
    print (savepath)
    carmudi = df.to_csv(savepath, index=None, header=False)

except:
    print ("Export exception:", e)
```

## Scraping Details from Price Price Auto

```python
browser = webdriver.Chrome(executable_path = filepath)
url_list = []
with open('urls_ppa.txt') as a:
    url_list = a.readlines()
url_list = [x.strip('\n') for x in url_list]
# print(url_list)
    
df=pd.DataFrame()
count=0
for url in url_list:
    try:
        browser.get(url)
        variants = browser.find_elements_by_class_name('s-variantTbl_trLink')
        for each in variants:
            try:
                name = each.find_element_by_class_name('variantTbl_name').text
                body = each.find_element_by_class_name('variantTbl_bodyTxt').text
                transmission = each.find_element_by_class_name('variantTbl_transmissionTxt').text
                fuel = each.find_element_by_class_name('variantTbl_fuelTxt').text
                df=df.append({'name':name,'body':body,'transmission':transmission,'fuel':fuel},ignore_index=True)
                df=df[['name','body','transmission','fuel']]
                count+=1
                print(count)
            except:
                print('End')
    except Exception:
        print("End")

savepath='C:\\Users\COBENA\Desktop\Learn\Code\FTW\Capstone Idea\Automotive\Price Price Auto\ppa_details.csv'
csv = df.to_csv(savepath,index=None)
browser.quit()
print('Finished Scraping:',count)
```
