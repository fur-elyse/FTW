```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import pandas as pd
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
filepath = os.getcwd()+'/geckodriver.exe'
print('Path to Driver is '+filepath)
```

```python
df = pd.read_csv('C:\\Users\COBENA\Desktop\Learn\Code\FTW\Capstone Idea\Automotive\Web Scraping/to_scrape_retail_price.csv',index_col=None)
df.head()
```

```python
browser=webdriver.Firefox(executable_path=filepath)
```

```python
model=[x for x in df['Model']]
brand=[x for x in df['Brand']]
count=0
skip=0
for x, y in zip(brand,model):
    x=str(x.strip().lower())
    y=str(y.strip().lower())
    browser.get('https://www.zigwheels.ph/new-cars/'+x+'/'+y+'/price')
    try:
        zig_srp=WebDriverWait(browser,5).until(EC.presence_of_element_located((By.XPATH,'/html/body/main/div[1]/div/div[1]/div[1]/ul/li[1]/div[2]'))).text
        zig_amt=WebDriverWait(browser,5).until(EC.presence_of_element_located((By.XPATH,'/html/body/main/div[1]/div/div[1]/div[1]/ul/li[3]/div[2]'))).text
        count+=1
        print(count)
        df=df.append({'zig_srp':zig_srp,'zig_amt':zig_amt,'x':x,'y':y},ignore_index=True)
        df=df[['x','y','zig_srp','zig_amt']]
    except:
        skip+=1
        print('skipped',skip)
print('done')
```
